```python
# Basic libraries
import pandas as pd                      # For data handling
import matplotlib.pyplot as plt          # For plotting

# Machine learning models and utilities
from sklearn.ensemble import RandomForestRegressor      # Base regressor
from sklearn.multioutput import MultiOutputRegressor    # For multi-step/multi-target prediction
from sklearn.metrics import mean_squared_error, r2_score  # Evaluation metrics

# Hyperparameter tuning using Bayesian Optimization
from skopt import BayesSearchCV                          # Performs efficient tuning
from skopt.space import Integer, Real                    # Defines search space for hyperparameters

# Cross-validation technique designed for time series data
from sklearn.model_selection import TimeSeriesSplit      # Preserves time order during CV

# For saving/loading trained models
import pickle

# Suppress warnings to keep output clean
import warnings
warnings.filterwarnings("ignore")

# Set plot style to dark theme
plt.style.use('dark_background')
```

```python
# Load the TCS stock data from a CSV file
tcs = pd.read_csv("..\data\TCS.csv")

# Check the number of rows and columns in the dataset
tcs.shape
```

```python
# Display the first 5 rows of the dataset to get a quick look at the data
tcs.head()
```

```python
# Convert the 'Date' column from string to datetime format for easier filtering and plotting
tcs['Date'] = pd.to_datetime(tcs['Date'])

# Create a boolean mask to filter data between Jan 2008 and Dec 2020
mask = (tcs['Date'] >= '2008-01-01') & (tcs['Date'] <= '2020-12-31')

# Apply the mask to get the filtered DataFrame
tcs_filtered = tcs.loc[mask]
```

```python
# Keep only the relevant columns for analysis and modeling
tcs = tcs_filtered[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
# Display the entire filtered and cleaned DataFrame
tcs
```

```python
# Plot the closing price of TCS over time

plt.figure(figsize=(12, 6))                             # Set the figure size
plt.plot(tcs['Date'], tcs['Close'], label='Close Price')  # Line plot of closing prices over time

plt.title("TCS Closing Price (2008–2020)")              # Plot title
plt.xlabel("Date")                                      # X-axis label
plt.ylabel("Close Price (INR)")                         # Y-axis label
plt.legend()                                            # Show legend
plt.tight_layout()                                      # Adjust layout to prevent overlap
plt.show()                                              # Display the plot
```

```python
# Create new columns for the next 5 days' closing prices
# This is useful for multi-step ahead forecasting
for i in range(1, 6):
    tcs[f'close_t+{i}'] = tcs['Close'].shift(-i)

# Drop the last 5 rows which now contain NaN values due to the shifting
tcs_model = tcs.dropna().copy()
```

```python
# Display the DataFrame after adding future close price columns and dropping NaNs
tcs_model
```

```python
# Features: Use current day's market data as input
X = tcs_model[['Open', 'High', 'Low', 'Close', 'Volume']]

# Targets: Predict closing prices for the next 5 days (multi-output regression)
y = tcs_model[['close_t+1', 'close_t+2', 'close_t+3', 'close_t+4', 'close_t+5']]
```

```python
# Time-based split (no shuffling) for training, validation, and testing
# This preserves the chronological order of stock prices

total = len(X)                             # Total number of samples

train_size = int(0.7 * total)              # 70% for training
val_size = int(0.15 * total)               # 15% for validation

# Split features and targets into train, validation, and test sets
X_train = X.iloc[:train_size]
y_train = y.iloc[:train_size]

X_val = X.iloc[train_size:train_size + val_size]
y_val = y.iloc[train_size:train_size + val_size]

X_test = X.iloc[train_size + val_size:]
y_test = y.iloc[train_size + val_size:]

# Display the number of samples in each set
print(f"Train size: {len(X_train)}, \nValidation size: {len(X_val)}, \nTest size: {len(X_test)}")
```

```python
# Initialize a MultiOutputRegressor with a base RandomForestRegressor
# This allows training a separate Random Forest model for each target (i.e., next 5 days' close prices)

model = MultiOutputRegressor(
    RandomForestRegressor(n_estimators=100, random_state=42)  # 100 trees, fixed seed for reproducibility
)
```

```python
# Train the multi-output Random Forest model on the training data
model.fit(X_train, y_train)
```

```python
# Predict the next 5 days' closing prices on the validation set
val_pred = model.predict(X_val)

print("Validation Set Performance:")
# Evaluate model performance for each future day separately
for i in range(5):
    mse = mean_squared_error(y_val.iloc[:, i], val_pred[:, i])  # Mean Squared Error
    r2 = r2_score(y_val.iloc[:, i], val_pred[:, i])              # R-squared score
    print(f"Day +{i+1} - MSE: {mse:.2f}, R²: {r2:.4f}")
```

```python
# Predict on the test set to evaluate model performance on unseen data
test_pred = model.predict(X_test)

print("\nTest Set Performance:")
# Evaluate each of the 5 future day predictions
for i in range(5):
    mse = mean_squared_error(y_test.iloc[:, i], test_pred[:, i])  # Test MSE
    r2 = r2_score(y_test.iloc[:, i], test_pred[:, i])              # Test R² score
    print(f"Day +{i+1} - MSE: {mse:.2f}, R²: {r2:.4f}")
```

```python
# Visualize actual vs predicted closing prices for Day t+1 on the test set

plt.figure(figsize=(12, 5))  # Set figure size

# Plot actual values for close_t+1
plt.plot(y_test.index, y_test['close_t+1'], label="Actual Close_t+1")

# Plot predicted values for close_t+1 (first column of test_pred)
plt.plot(y_test.index, test_pred[:, 0], label="Predicted Close_t+1", color="red")

plt.title("Actual vs Predicted Close Price (t+1) – Test Set")  # Chart title
plt.xlabel("Time Index")                                       # X-axis label
plt.ylabel("Close Price")                                      # Y-axis label
plt.legend()                                                   # Show legend
plt.tight_layout()                                             # Optimize spacing
plt.show()                                                     # Display plot
```

# Hyper Parameter Tuning (Bayesian)

```python
# Define the hyperparameter search space for Bayesian Optimization
# These will be tuned for the base RandomForestRegressor inside MultiOutputRegressor

param_space = {
    'estimator__n_estimators': Integer(50, 300),          # Number of trees in the forest
    'estimator__max_depth': Integer(5, 30),               # Maximum depth of each tree
    'estimator__min_samples_split': Integer(2, 10),       # Minimum samples required to split an internal node
    'estimator__min_samples_leaf': Integer(1, 5),         # Minimum samples required at each leaf node
    'estimator__max_features': Real(0.5, 1.0, prior='uniform')  # Fraction of features to consider when looking for the best split
}
```

```python
# Prepare model and cross-validator for Bayesian optimization

# Initialize the base Random Forest Regressor
base_rf = RandomForestRegressor(random_state=42)

# Wrap it with MultiOutputRegressor to handle multi-step forecasting
multi_rf = MultiOutputRegressor(base_rf)

# Use TimeSeriesSplit for cross-validation that respects time order
tscv = TimeSeriesSplit(n_splits=3)
```

```python
# Set up Bayesian Optimization using BayesSearchCV

opt = BayesSearchCV(
    estimator=multi_rf,           # Multi-output Random Forest model
    search_spaces=param_space,    # Hyperparameter space to explore
    cv=tscv,                      # Time series cross-validation strategy
    n_iter=25,                    # Number of optimization iterations
    scoring='r2',                 # Use R² score to evaluate model performance
    verbose=1,                    # Show progress during training
    n_jobs=-1,                    # Use all available CPU cores
    random_state=42              # For reproducibility
)
```

```python
# Combine training and validation sets for final tuning
# This gives the model more data to generalize better during hyperparameter search

X_tune = pd.concat([X_train, X_val])
y_tune = pd.concat([y_train, y_val])

# Check the shape of the combined dataset
print(X_tune.shape, y_tune.shape)
```

```python
# Run Bayesian hyperparameter search on the combined training + validation set
opt.fit(X_tune, y_tune)
```

```python
# Display the best hyperparameters found during Bayesian optimization
print("Best Parameters:", opt.best_params_)
```

```python
# Retrieve the best model found by Bayesian search (already trained on X_tune)
best_model = opt.best_estimator_
```

```python
# Make predictions on the test set using the best tuned model
y_test_pred = best_model.predict(X_test)
```

```python
# Evaluate the tuned model's performance on the test set

print("\n Test Set Performance After Bayesian Tuning:")
for i in range(5):
    mse = mean_squared_error(y_test.iloc[:, i], y_test_pred[:, i])  # Mean Squared Error
    r2 = r2_score(y_test.iloc[:, i], y_test_pred[:, i])              # R² Score
    print(f"Day +{i+1} - MSE: {mse:.2f}, R²: {r2:.4f}")
```

```python
# Plot actual vs predicted closing price for Day t+1 using the tuned model

plt.figure(figsize=(12, 5))  # Set figure size

# Plot actual values
plt.plot(y_test.index, y_test['close_t+1'], label="Actual Close_t+1")

# Plot predicted values from the tuned model
plt.plot(y_test.index, y_test_pred[:, 0], label="Predicted Close_t+1", color="red")

plt.title("Actual vs Predicted Close Price (t+1) – Test Set")  # Title
plt.xlabel("Time Index")                                       # X-axis label
plt.ylabel("Close Price")                                      # Y-axis label
plt.legend()                                                   # Add legend
plt.tight_layout()                                             # Adjust spacing
plt.show()                                                     # Display the plot
```

```python
# Save the model
with open('multioutput model.pkl', 'wb') as file:
    pickle.dump(best_model, file)

# # Load the model
# with open('model.pkl', 'rb') as file:
#     model = pickle.load(file)
```
